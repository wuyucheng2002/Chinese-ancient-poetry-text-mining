```python
import re
done_df = []
none_df = []

for i, row in df.iterrows():
    if i % 1000 == 0:
        print(i)
    content = row['内容']
    pattern = re.compile('（[^）]*）')
    content = pattern.sub('', content)
    
    words = list(tokenizer.tokenize(content))
    words = [c for c in words if c not in punc]
    
    text = ' '.join(words)
    
    if not pd.isna(row['主题']):
        done_df.append({
            '诗词id': row['诗词id'], 
            '内容': text, 
            '主题': row['诗词id']
        })
    else:
        none_df.append({
            '诗词id': row['诗词id'], 
            '内容': text
        })
    
done_df = pd.DataFrame(done_df)
none_df = pd.DataFrame(none_df)
```

```python
done_df
```

```python
none_df
```

```python
X = done_df['内容']
y = done_df['主题']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
print(len(X_train), len(X_test))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
tf_vec = CountVectorizer(max_df=0.4, max_features=30000)
X_train_tf = tf_vec.fit_transform(X_train)
X_test_tf = tf_vec.transform(X_test)
print(X_train_tf.shape, X_test_tf.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train_tf, y_train)
```

```python
y_hat_test = logreg.predict(X_test_tf)
# print(round(logreg.score(X_test_tf, y_test), 5))
print(accuracy_score(y_test, y_hat_test))
print(classification_report(y_test, y_hat_test))
```

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=1000)
new_X_train_tf = svd.fit_transform(X_train_tf)
new_X_test_tf = svd.transform(X_test_tf)
```

```python
print(sum(svd.explained_variance_ratio_[:]))
```

```python
logreg2 = LogisticRegression()
logreg2.fit(new_X_train_tf, y_train)
```

```python
y_hat_test2 = logreg2.predict(new_X_test_tf)
print(accuracy_score(y_test, y_hat_test2))
print(classification_report(y_test, y_hat_test2))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1000)
new_X_train_tf = pca.fit_transform(X_train_tf.todense())
new_X_test_tf = pca.transform(X_test_tf.todense())
```

```python
print(sum(pca.explained_variance_ratio_[:1000]))
```

```python
logreg2 = LogisticRegression()
logreg2.fit(new_X_train_tf, y_train)
```

```python
y_hat_test2 = logreg2.predict(new_X_test_tf)
print(accuracy_score(y_test, y_hat_test2))
print(classification_report(y_test, y_hat_test2))
```
